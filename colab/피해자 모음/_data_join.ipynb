```python
import pandas as pd
```

```python
import os
from glob import glob
```

```python
def change_date(df_data):
  return pd.to_datetime(df_data.map(lambda x : f'{x[:4]}-{x[4:6]}-{x[6:8]}'))
```

```python
df_1 = pd.read_csv('/content/drive/MyDrive/Project_2/Data/피해자 모음/금융사기 피해자 지역 데이터/TCH_32_20220131_C_001.csv')
```

```python
df_1
```

```python
file_join_1 = os.path.join('/content/drive/MyDrive/Project_2/Data/피해자 모음/금융사기 피해자 지역 데이터', 'TCH_32_2*.csv')
list_files_1 = glob(file_join_1)
```

```python
columns_name = ['WDAR_CTPR_NM', 'LEGAL_SIGUNGU_NM', 'RGSTN_DT']
df_1 = pd.DataFrame(columns=columns_name)

for file in list_files_1:
  f = pd.read_csv(file, names=columns_name, index_col=0)
  df_1 = pd.concat([df_1, f], ignore_index = True)
```

```python
df_1['RGSTN_DT'] = change_date(df_1['RGSTN_DT'])
```
