```python
df3['RGSTN_DT'] = pd.to_datetime(df3['RGSTN_DT'])
```

```python
from datetime import datetime
```

```python
df3.loc[(df3['IBBN'] == '나이 물품') & (df3['RGSTN_DT'] >= datetime(2021,1,1)) & (df3['RGSTN_DT'] <= datetime(2021,10,1)), 'DAMAGE'].shape
```

```python
def set_weeks(df):
    weeks_list = ['월', '화', '수', '목', '금', '토', '일']
    weeks = df['RGSTN_DT'].dt.dayofweek
    df['weeks'] = weeks.map(lambda x : weeks_list[x])
    df['is_week'] = weeks.map(lambda x: '평일' if x < 5 else '주말')

    return df
```

```python
df3['RGSTN_DT'] = pd.to_datetime(df3['RGSTN_DT'])
```

```python
df3 = set_weeks(df3)
```
