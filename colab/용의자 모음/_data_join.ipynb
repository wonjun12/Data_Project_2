```python
def change_date(df_data):
  return pd.to_datetime(df_data.map(lambda x : f'{x[:4]}-{x[4:6]}-{x[6:8]}'))
```

```python
file_join_1 = os.path.join('/content/drive/MyDrive/Project_2/Data/용의자 모음/금융사기 용의자 지역 데이터', 'TCH_33_2*.csv')
list_files_1 = glob(file_join_1)
```

```python
columns_name = ['WDAR_CTPR_NM', 'LEGAL_SIGUNGU_NM', 'RGSTN_DT']
df_1 = pd.DataFrame(columns=columns_name)

for file in list_files_1:
  f = pd.read_csv(file, names=columns_name)
  df_1 = pd.concat([df_1, f], ignore_index = True)
```

```python
df_1['RGSTN_DT'] = change_date(df_1['RGSTN_DT'])
```

```python
df_1 = df_1.sort_values('RGSTN_DT').reset_index(drop='index')
```

```python
df_1.to_csv('/content/drive/MyDrive/Project_2/Data/용의자 모음/TCH_Suspect_Resion.csv', index=False, encoding='utf-8-sig')
```

```python
df_1_back = df_1.groupby(['RGSTN_DT', 'WDAR_CTPR_NM'])['LEGAL_SIGUNGU_NM'].value_counts().to_frame()
```

```python
df_1_back
```

```python
df_1_r = pd.DataFrame(
    data=[[data[0], data[1], data[2]] for data in df_1_back.index],
    columns=['RGSTN_DT','WDAR_CTPR_NM', 'LEGAL_SIGUNGU_NM']
)
```

```python
df_1_r['COUNT'] = df_1_back.values
```

```python
df_1_r.to_csv('/content/drive/MyDrive/Project_2/Data/용의자 모음/금융사기 용의자 지역 데이터/TCH_33.csv', index=False, encoding='utf-8-sig')
```
